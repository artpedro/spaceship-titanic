```python
data = np.load('../data/train_curated.npz',allow_pickle=True)
X = data['train']
Y = data['target']
feats_names = data['feats']

data = np.load('../data/test_curated.npz',allow_pickle=True)
X_test = data['test']
LABELS_SUB = data['labels']
print(X.shape)
print(X_test.shape)
```

```python
all_data_train = True

x_train, x_cv, y_train, y_cv = train_test_split(X,Y,test_size=0.3)

if all_data_train:
    x_train = X
    y_train = Y

x_train.shape
```

```python
pipeline_optimizer = TPOTClassifier()
pipeline_optimizer = TPOTClassifier(generations=25, population_size=100, cv=5,
                                    random_state=42, verbosity=2)
```
